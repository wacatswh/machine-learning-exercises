```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering
import scipy.cluster.hierarchy as sch

countries_df = pd.read_csv('Dataset/countries_geocodes.csv')

x = countries_df[['latitude', 'longitude']]
```

# Part 1: K-Means Clustering

```python
# k = 5
kmeans = KMeans(n_clusters = 5, random_state = 8)

kmeans.fit(x)

y_kmeans = kmeans.fit_predict(x)

plt.figure(figsize = (12, 12))

cluster0 = plt.scatter(x.iloc[y_kmeans == 0, 0], x.iloc[y_kmeans == 0, 1], s = 100, \
                       c = 'red', label = 'Cluster 1')
cluster1 = plt.scatter(x.iloc[y_kmeans == 1, 0], x.iloc[y_kmeans == 1, 1], s = 100, \
                       c = 'blue', label = 'Cluster 2')
cluster2 = plt.scatter(x.iloc[y_kmeans == 2, 0], x.iloc[y_kmeans == 2, 1], s = 100, \
                       c = 'green', label = 'Cluster 3')
cluster3 = plt.scatter(x.iloc[y_kmeans == 3, 0], x.iloc[y_kmeans == 3, 1], s = 100, \
                       c = 'pink', label = 'Cluster 4')
cluster4 = plt.scatter(x.iloc[y_kmeans == 4, 0], x.iloc[y_kmeans == 4, 1], s = 100, \
                       c = 'purple', label = 'Cluster 5')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, \
            c = 'yellow', label = 'Centroids')

plt.legend((cluster0, cluster1, cluster2, cluster3, cluster4),
          ('Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4', 'Cluster 5'),
          scatterpoints = 1,
          loc = 'upper left',
          ncol = 3,
          fontsize = 8)

plt.title('Latitude vs Longitude')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.show()

print("K = 5 - Sum of squared distances to centroids:", kmeans.inertia_)
```

# Part 3: Identify the best continent clustering

```python
# Cluster mapping
cluster_map = {
    0: 'North America',
    1: 'Africa',
    2: 'Asia',
    3: 'Europe',
    4: 'Australia',
    5: 'Asia',
    6: 'Europe',
    7: 'Africa',
    8: 'South America',
    9: 'Asia',
    10: 'Antarctica'
}
cluster_list = list()

for cluster_label in cluster:
    cluster_list.append(cluster_map.get(cluster_label))
```

```python
# Combine countries_df and cluster_list labels together into final_cluster
final_cluster = pd.concat([countries_df, pd.DataFrame(cluster_list, columns = ['cluster'])], axis = 1)
final_cluster
```
